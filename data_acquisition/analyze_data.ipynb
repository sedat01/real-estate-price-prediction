```python
import pandas as pd
import seaborn as sns 
pd.set_option("display.max_columns", 100)
pd.set_option('display.max_colwidth', -1)
data = pd.read_csv("collected1.csv",low_memory=False)
data
```

```python
from pandas import Categorical

data["energy_heatingType"]=data["energy_heatingType"].fillna("other")
data['energy_heatingType'] = data['energy_heatingType'].astype('category')
data.drop("Unnamed: 0", axis=1,inplace=True)

data
```

```python
from matplotlib.pyplot import hist



sns.displot(data["price"])
```

```python
from matplotlib.pyplot import xlabel, ylabel
import matplotlib.ticker as ticker

data =data[data["bedroom_count"]<=6]
data =data[data["bedroom_count"]>1]
data["bedroom_count"] = data["bedroom_count"].astype(int)
data = data[data["certificates_primaryEnergyConsumptionLevel"]<=2000]
data = data[data["subtype"] == "house"]
data.drop(["Surroundings type","Primary energy consumption","Heating type"], axis=1,inplace=True,errors="ignore")

from pandas.api.types import is_numeric_dtype  
def convert_cats(df):
        cats = []
        for col in df.columns:
            if is_numeric_dtype(df[col]):
                pass
            else:
                cats.append(col)
        for col in cats:
            df[col] = df[col].astype('category')
convert_cats(data)

data = data[data["land_surface"] <= 5000]
scatter = sns.scatterplot(data=data, x="certificates_primaryEnergyConsumptionLevel", y="price")
scatter.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{:,.0f}'.format(y/1000) + 'K'))
scatter.set(xlabel="Energy consumption per square meter", ylabel="Average price (Euros)")
```

```python
data = data[data["Living area"]<1000]
antwerpen = data[data.Province.isin (["Namur","Brussels"])]
sns.relplot(data=data, x="Living area", y="price")
```

```python


labels = ["Needs renovation","Good","As new"]
barchart = sns.barplot(data=data,x="bedroom_count",y="price", hue="building_condition", errorbar=None)
h, l = barchart.get_legend_handles_labels()
barchart.legend(h, labels, title="Bulding Condition")
barchart.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{:,.0f}'.format(y/1000) + 'K'))
barchart.set(xlabel="Bedrooms", ylabel="Average price (Euros)")
barchart
```
